```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

data['rate'] = data['rate'].apply(handleRate)
```

```python
data
```

```python
#Q1:What type of do the majority of customers order from?
```

```python
sns.countplot(x='listed_in(type)',data = data)
```

```python
#conclusion: Majority order is dining
```

```python
data
```

```python
#Q2: How many votes has each type of restaurant received from customers?
```

```python
sns.catplot(x='listed_in(type)',y='votes',data = data,hue='listed_in(type)',kind = 'bar')
```

```python
#conclusion: others received the greater number of votes
```

```python
#Q3: what are the ratings that the majority of restaurant have received?
```

```python
sns.histplot(x='rate',data=data)
```
